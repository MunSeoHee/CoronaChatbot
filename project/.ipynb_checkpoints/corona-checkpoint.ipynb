```python
import pandas as pd
import numpy as np
seoul_data=pd.read_csv("seoul_corona.csv")
```

```python
seoul_data= seoul_data.drop("Unnamed: 0",axis=1)
```

```python
seoul_data= seoul_data.drop("성별",axis=1)
seoul_data= seoul_data.drop("나이",axis=1)
seoul_data= seoul_data.drop("증상 의심 날짜",axis=1)
```

```python
seoul_data=seoul_data.sort_values(by="확진 날짜", ascending=False)
```

```python
seoul_data=seoul_data.replace("X","확인불가")
```

```python
seoul_data.to_csv("seoul.csv", encoding="utf-8-sig")
```
